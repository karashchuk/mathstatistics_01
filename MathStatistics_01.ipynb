###  Задача 1

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.


Количество исходов когда хотя бы один туз из 4 выбранных карт = количество исходов когда выбран 1 туз + количество исходов когда 2 туза + количество исходов когда 3 туза + количество исходов когда 4 туза

Количества исходов: 

$q(1A) = C_4^1 * C_{51}^3$

$q(2A) = C_4^2 * C_{50}^2$

$q(3A) = C_4^3 * C_{49}^1$

$q(4A) = C_4^4 * C_{48}^0 = 1*1 = 1$

И общее количестов исходов $q(A) = q(1A) + q(2A) + q(3A) + q(4A) $

```python
import math

def comb(n,k):
    return int(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))

q = comb(4,1)*comb(51,3) + comb(4,2)*comb(50,2)  + comb(4,3)*comb(49,1)  + comb(4,4)*comb(48,0) 
print(f'Общее количество исходов при которых из 4 выбранных карт будет хотя бы один туз = {q} \
(из всех возможных {comb(52,4)} вариантов)')
```

### Задача 2

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?

Всего возможных рассадок - это количество перестановок:  $n! = 7!$

2 человека рядом могут сесть 2 мя способами

всего из 7 мест 6 вариантов посадить пару

остальные 5 чедловек могут сесть $5!$ способами,

тогда верятность посадить рядом двух определенных человек: 

$$P(2) = \frac{2*6*5!}{7!} = \frac{2*6}{6*7} = \frac{2}{7}$$

для 3 человек рядом:

количество общее воможности рассадить = $7!$

3 человека рядом могут сесть $3!$ способами

5 возможных вариантов для 3 человек сесть рядом

остальные 4 человека могут сесть $4!$ способами

тогда верятность посадить рядом двух определенных человек: 

$$P(3) = \frac{3!*5*4!}{7!} = \frac{3!*5}{5*6*7} = \frac{6}{6*7}= \frac{1}{7}$$

### Задача 4

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?


$P(A) = \frac{1}{2}$, так как это вероятность выпадения {2,4,6} из {1,2,3,4,5,6}

$P(B) = \frac{1}{2}$, так как это вероятность выпадения {4,5,6} из {1,2,3,4,5,6}

Для того чтобы события A и B были независимы, должно быть справедливо равенство:

$$P(A|B) = P(A)$$

Но при наступлении события B вероятность наступления события А будет $\frac{2}{3}$, так как среди трех чисел больше трех - 2 будут четные, то есть 

$$P(A|B) = \frac{2}{3} \,\,\, \neq \,\,\, P(A) = \frac{1}{2}$$

то есть события A и B - зависимы

### Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

_Вопрос_: какова вероятность, что вы действительно больны ей?

_Подсказка_: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Пусть событие $B$ значит что человек болеет

событие $\overline B$ значит что человек здоров

тогда:
$$P(B) = 0.001$$
$$P(\overline B) = 0.999$$

Событие $A$ -  это положительный тест

тогда

$$P(A|B) = 0.99 $$
$$P(A|\overline B) = 0.01$$

Тогда вероятность что человек болен при положительном тесте по формуле Байеса:

$$P(B|A) = \frac{P(A|B)\cdot P(B) }{P(A)}  = \frac{P(A|B)\cdot P(B) }{P(A|B)\cdot P(B) + P(A|\overline B)\cdot P(\overline B)} = \frac{0.99 * 0.001}{0.99 * 0.001 + 0.01*0.999} = 0.090163934 \approx 0.09 $$




То есть вероятность того что человек болен на самом деле при положительном тесте чуть более $9\,\%$
